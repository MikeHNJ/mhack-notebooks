# Plot a message from data points
This example plots a message from data points saved in a CSV. 
The CSV file was generated from an image file using the
[create-message-data notebook](create-message-data.ipynb).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load from CSV

```python
message_points = pd.read_csv( "C:\\Users\\micha269\\pydata-book\\mike\\message_points.csv")
```

## Sample the data and create new columns for plotting

```python
sample_points = message_points.sample(n=2000)
rows, cols = sample_points.shape
sample_points['adjusted_x'] = sample_points['x'] + np.random.normal(loc=0.0, scale=1, size=(rows))
sample_points['adjusted_y'] = sample_points['y'] + np.random.normal(loc=0.0, scale=1, size=(rows))
sample_points
```

```python
sns.scatterplot('adjusted_x', 'adjusted_y', data=sample_points, hue='adjusted_y')
```
